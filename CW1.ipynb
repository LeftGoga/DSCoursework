```python
#1E
stud_age["birthday"]=pd.to_datetime(stud_age["birthday"], errors='coerce')
stud_age_nona=stud_age["birthday"].dropna()

def age(birthdate):
    
    today = date.today()
    
    one_or_zero = ((today.month, today.day) < (birthdate.month, birthdate.day))
    
    year_difference = today.year - birthdate.year
    
    age = year_difference - one_or_zero
    
    return age


list1=[]
for i in stud_age_nona:
    list1.append(age(i))
stud_age_nona=stud_age_nona.to_frame()
stud_age_nona.insert(loc=0, column="age", value=list1)


stud_age_nona=stud_age_nona.drop(stud_age_nona[stud_age_nona["age"]<0].index)

#вычисление нужных величины
print(max(stud_age_nona["age"]))
print(min(stud_age_nona["age"]))
print(np.mean(stud_age_nona["age"]))
print(np.median(stud_age_nona["age"]))
```

```python
#1
big_df["start_date"]=pd.to_datetime(big_df["start_date"])
increase=big_df[["title","module_number","is_homework","module_title","lesson_number","start_date"]].sort_values("start_date")
increase["date"]=pd.to_datetime(increase['start_date']).dt.to_period('M')
increase=increase[increase["is_homework"]==True]

increase1=increase[["title","module_number"]]

increase2=increase[["title","module_number","date"]]


increase1=increase1[["title","module_number",]].groupby(["title","module_number"]).agg("min")

i_merged=pd.merge(increase1,increase2,how="inner",on=["title","module_number"])
i_merged["student_counter"]=1
i_counted=i_merged[["title","date","student_counter"]].groupby(["title","date"]).count().rename(columns={"student_counter":"count"}).reset_index()



i_counted
```

```python
#1
conversion1=big_df[["title","module_number", "status","is_homework","module_title","field","lesson_number"]]
conversion2=big_df[["title","module_number", "status","is_homework","module_title","field","lesson_number"]]
conversion1=conversion1[conversion1["is_homework"]==True]

conversion1=conversion1[["title","module_number","is_homework","module_title","field","lesson_number"]].groupby(["title","module_number"]).agg("min")

c_merged=pd.merge(conversion1,conversion2,how="inner",on=["title","module_number","is_homework","module_title","field","lesson_number"])

c_merged


c_merged["counter"]=1

conv_done=c_merged.loc[c_merged["status"]=="done"][["title","module_number","module_title","field","counter"]].groupby(["title","module_number","module_title","field"]).count()
conv_done=conv_done.rename(columns={"counter":"counter_done"})
conv_total=c_merged[["title","module_number","module_title","field","counter"]].groupby(["title","module_number","module_title","field"]).count()

conv_total=conv_total.rename(columns={"counter":"counter_total"})



conv_counted=pd.merge(conv_done,conv_total,left_index=True,right_index=True) 

titles=set(conv_counted.reset_index()["title"])

conv_final=pd.DataFrame()

for i in titles:
    temp=pd.DataFrame()
    temp=conv_counted[conv_counted.index.get_level_values("title")==i]
    temp["prev_done"]=temp["counter_done"].shift(1)


    temp["conversion"]=temp["counter_total"]/temp["prev_done"]
    l=[conv_final,temp]
    conv_final=pd.concat(l)
conv_final

```

```python
#2
fig,ax=plt.subplots(nrows=15,ncols=1, figsize=(10,50),constrained_layout = True)
conv_final=conv_final.reset_index()
titles=set(conv_final["title"])

for i,title in enumerate(titles):
    data=conv_final[conv_final["title"]==title][["module_number","conversion"]]
    x_range=range(len(data))
    axs=ax[i]

    axs.bar(x=x_range,height=data["conversion"])
    axs.set_title(title)
    axs.set_ylabel("конверсия")
    axs.set_xlabel("номер модуля")
    axs.set_xticks(x_range)
    axs.set_xticklabels(data["module_number"])
```
